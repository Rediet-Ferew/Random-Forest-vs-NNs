```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.preprocessing import QuantileTransformer

class SklearnMLPWrapper:
    def __init__(self, params, args):
        # Initialize the sklearn MLP model
        self.model = MLPClassifier(
            hidden_layer_sizes=(params["hidden_dim"],) * params["n_layers"], 
            activation='relu',
            solver='adam',
            alpha=0.0001,
            learning_rate_init=params["learning_rate"],
            max_iter=1,  
            random_state=42
        )
        # Initialize the QuantileTransformer
        self.scaler = QuantileTransformer(output_distribution='normal', random_state=42)

    def fit(self, X, y, X_val=None, y_val=None, epochs=300, batch_size=256):
        # Apply QuantileTransformer to the features
        X = self.scaler.fit_transform(X)
        if X_val is not None:
            X_val = self.scaler.transform(X_val)

        # Fit the model manually for each epoch
        for epoch in range(epochs):
            self.model.set_params(max_iter=epoch + 1)  # Increase iterations each epoch
            self.model.fit(X, y)  # sklearn handles batching internally during fitting

            # Compute and print training accuracy
            if X_val is not None and y_val is not None:
                val_preds = self.predict_helper(X_val)
                val_accuracy = accuracy_score(y_val, val_preds)
                print(f'Epoch {epoch + 1}/{epochs}, Validation Accuracy: {val_accuracy}')

    def predict_helper(self, X):
        return self.model.predict(X)

    def predict(self, X_test):
        # Apply the same QuantileTransformer to the test data
        X_test = self.scaler.transform(X_test)
        return self.predict_helper(X_test)
```
